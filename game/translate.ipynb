```python
# draw f(x) = w1 * x1 + w2 * x2 + b in pyplot
import numpy as np
import matplotlib.pyplot as plt

# define the line
w1 = 0.5
w2 = 1.0
b = -2.0

# create 2 points
x1 = np.array([-2, 4])
x2 = (-b - w1 * x1) / w2

# plot the line
plt.plot(x1, x2, 'k-')
plt.axis([-3, 5, -3, 5])
plt.show()

```
