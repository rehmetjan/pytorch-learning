```python
import numpy as np
import matplotlib.pyplot as plt

x1 = 1
y1 = 2 * x1

x2 = 3
y2 = 2 * x2

xs = [x1, x2]
ys = [y1, y2]

plt.scatter(x1, y1)
plt.scatter(x2, y2)

plt.plot(xs, ys)
plt.show()

```
