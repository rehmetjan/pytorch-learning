```python
import sys, os

def get_beans(counts):
    xs = np.random.rand(counts)
    # xs = np.sort(xs)
    # ys = [1.7 * x + np.random.rand() / 5 for x in xs]
    ys = [1.2 * x + np.random.rand() / 10 for x in xs]
    # ys = np.random.rand(counts)
    return xs *2, ys

xs, ys = get_beans(100)
print(xs)
print(ys)
plt.scatter(xs, ys)
plt.show()
```

```python
def make_beans(number):
    Xs = np.random.rand(number)
    Xs = Xs.round(2) * 10
    # Xs = Xs * 10
    print(Xs)
    for x in Xs:
        print(x, type(x))
    # Ys = [np.random.rand(number)]
    Ys = [1.2 * x + np.random.rand() / 10 for x in Xs]
    for y in Ys:
        print("y:", y)

    return Xs, Ys

Xs, Ys = make_beans(10)

print(Xs)
print(Ys)

plt.scatter(Xs[-1], Ys[-1])
```
