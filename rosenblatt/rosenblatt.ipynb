```python
import numpy as np
import matplotlib.pyplot as plt

def make_beans(number):
    xs = np.random.rand(number)
    ys = [1.2 * x + np.random.rand() / 5 for x in xs]
    return xs, ys

number_of_beans = 50
xs, ys = make_beans(number_of_beans)

plt.scatter(xs, ys)
plt.show()
```

now we add learning rate to this `w`

$y_p  =  w + \alpha ex$

```python
w = 0.5
alpha = 0.05

for i in range(number_of_beans):
    for j in range(number_of_beans):
        x = xs[j]
        y = ys[j]
        y_pred = w * x
        e = y - y_pred
        w = w + alpha * e * x
        print(w)

y_pred = w * xs
plt.scatter(xs, ys)
plt.plot(xs, y_pred, color='red')
plt.show()
```
