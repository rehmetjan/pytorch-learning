```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# make beans function based on number of given beans
def get_beans(number):
    xs = np.random.rand(number)
    ys = [1.2 * x + np.random.rand() / 5 for x in xs]
    return xs, ys
```

```python
beans_count = 100
xs, ys = get_beans(beans_count)
plt.scatter(xs, ys)
```

```python
# w
w = 0.1
y_pre  = w * xs
plt.scatter(xs, ys)
plt.plot(xs, y_pre, color='red')
plt.show()
```

```python
# es = (ys - y_pre) ** 2
# sum_e = np.sum(es) * (1 / beans_count)
# print(sum_e)
ws = np.arange(0, 3, 0.1)
print(ws)

es = []
for w in ws:
    y_pre = w * xs
    e = (1 / beans_count) * np.sum((ys - y_pre) ** 2)
    # print("w: " + str(w) + " e: " + str(e))
    es.append(e)
print(es)
plt.scatter(xs, ys)
plt.plot(ws, es)
```

```python
w_min = np.sum(xs * ys) / np.sum(xs * xs)
print(w_min)
plt.scatter(xs, ys)
plt.plot(xs, y_pre, color='red')
plt.show()
```

```python
y_pre = w_min * xs
plt.scatter(xs, ys)
plt.plot(ws, es)
plt.plot(xs, y_pre, color='red')
```
